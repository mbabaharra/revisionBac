# Structure de donn√©es arborescentes : les arbres binaires

## Rappel de cours et exemples



#### Vocabulaire
‚Ä¢	Chaque √©l√©ment de l'arbre est appel√© **n≈ìud** (par exemple : A, B, C, D,...,P et Q sont des n≈ìuds)  
‚Ä¢	Le nombre de n≈ìud de l'arbre est appel√©e **ordre** (L'arbre exemple est d'ordre 17, on parle aussi de **taille**).  
‚Ä¢	Le **n≈ìud initial** (A) est appel√© n≈ìud **racine** ou plus simplement racine  
‚Ä¢	On dira que le n≈ìud E et le n≈ìud D sont les **fils** (ou les **enfants**) du n≈ìud B. On dira que le n≈ìud B est le **p√®re** des n≈ìuds E et D  
‚Ä¢	Le **degr√© d'un sommet** correspond au **nombre de fils** de ce sommet  
‚Ä¢	Un n≈ìud n'ayant **aucun fils** est appel√© **feuille** (exemples : D, H, N, O, J, K, L, P et Q sont des feuilles)  
‚Ä¢	√Ä partir d'un n≈ìud (qui n'est pas une feuille), on peut d√©finir un **sous-arbre gauche** et un **sous-arbre droit**  
‚Ä¢	On appelle **ar√™te** (ou **arcs**) le segment qui relie 2 n≈ìuds.  
‚Ä¢	Une branche est la suite de n≈ìud qui part de la racine vers une feuille. (ex : ABD ou ACGMQ etc.)  
‚Ä¢	On appelle **profondeur** d'un n≈ìud ou d'une feuille dans un arbre binaire le **nombre d'ar√™tes** du chemin qui va de la racine √† ce n≈ìud.   
‚Ä¢	On appelle **hauteur** d'un arbre la **profondeur maximale** des n≈ìuds de l'arbre. 

#### Parcourir un arbre en profondeur (DFS)



Trois parcours d'arbre sont √† connaitre :  
* l‚Äôordre **pr√©fixe** : on liste chaque sommet la premi√®re fois qu‚Äôon le rencontre dans la balade. Ce qui donne ici : `r, a, c, h, d, i , j ,l, b, e, k, f`   
* l‚Äôordre **suffixe (ou postfixe)** : on liste chaque sommet la derni√®re fois qu‚Äôon le rencontre. Ce qui donne ici : `h, c, i ,l, j ,d, a, k, e, f ,b, r`   
* l‚Äôordre **infixe** : on liste chaque sommet ayant un fils gauche la seconde fois qu‚Äôon le voit et chaque sommet sans fils gauche la premi√®re fois qu‚Äôon le voit. Ce qui donne ici : `c, h, a, i ,d, l, j , r, k, e b, f` 


#### Parcourir un arbre en largeur (BFS)



Le parcours en largeur consiste √† parcourir tous les n≈ìuds situ√©s √† la m√™me profondeur de gauche √† droite avant de passer aux n≈ìuds de la profondeur suivante.

## Algorithmes √† conna√Ætre sur les arbres binaires

√Ä chaque n≈ìud d'un arbre binaire, on associe une cl√© ("valeur" associ√©e au n≈ìud), un "sous-arbre gauche" et un "sous-arbre droit"  
On remarque que les arbres sont des structures qui peuvent √™tre d√©crites de mani√®re r√©cursive. Le cas de base serait un arbre ne contenant qu'un seul n≈ìud (une feuille)  
le cas g√©n√©ral correspond √† un arbre contenant au moins un sous-arbre, ce dernier peut alors de nouveau √™tre d√©crit en utilisant cette description.  

### Hauteur d'un arbre  



On appelle hauteur d'un arbre la profondeur maximale des n≈ìuds de l'arbre.  
On peut d√©finir la hauteur r√©cursivement sur la structure de l'arbre de la mani√®re suivante :  
‚Ä¢	l'arbre vide √† une hauteur de 0  
‚Ä¢	un arbre non vide a pour hauteur le maximum des hauteurs de ses deux sous-arbres, auquel on ajoute 1  

L'impl√©mentation des algorithmes sur les arbres binaires d√©pend de la mani√®re choisie pour repr√©senter l'arbre lui-m√™me. Dans le cas de cet exemple, j'ai choisi de repr√©senter les branches vides par des listes vides.  

```
arbreL= ['A', ['B',['E',[],[] ]  ,['F',[],[]]  ],  ['D',  ['G', [],[] ]  ,[] ] ]

def hauteur_arbre_liste(arbre):
    if arbre[1]==[] and arbre[2]==[]:
        return 0
    elif arbre[1]==[]:
        return 1 + hauteur_arbre_liste(arbre[2])
    elif arbre[2]==[]:
        return 1 + hauteur_arbre_liste(arbre[1])
    else :
        return 1 + max(hauteur_arbre_liste(arbre[1]),hauteur_arbre_liste(arbre[2]))
```
Ainsi :  `arbre[0]`: racine ; `arbre[1]` = sous arbre gauche ; `arbre[2]` = sous arbre droit  

Le cas terminal si deux branches vides la hauteur est alors de 0  
Le cas g√©n√©ral :  
-	si la branche de gauche est vide, on relance la fonction sur la branche de droite, ce qui signifie que l'on peut aller plus loin dans l'arbre donc on ajoute 1  
-	si la branche de droite est vide, on relance la fonction sur la branche de gauche, ce qui signifie que l'on peut aller plus loin dans l'arbre donc on ajoute 1  
-	si aucune des deux branches n'est vide, on relance la fonction sur les deux branches en r√©cup√©rant la hauteur max entre les deux branches  

### Taille d'un arbre  

La taille (ou ordre) d‚Äôun arbre correspond √† son nombre de n≈ìuds  
On peut d√©finir la taille r√©cursivement sur la structure de l'arbre de la mani√®re suivante :  
‚Ä¢	l'arbre vide √† une taille de 1 (un n≈ìud racine mais aucun sous-arbre)  
‚Ä¢	un arbre non vide contient donc au moins 1 n≈ìud (sa racine) auquel on va ajouter la taille des sous-arbre gauche et/ou droit s'ils existent.  

```
def taille_arbre_liste(arbre):
    if arbre[1]==[] and arbre[2]==[]:
        return 1
    elif arbre[1]==[]:
        return 1 + taille_arbre_liste(arbre[2])
    elif arbre[2]==[]:
        return 1 + taille_arbre_liste(arbre[1])
    else :
        return 1 + taille_arbre_liste(arbre[1])+ taille_arbre_liste(arbre[2])
```

Cas terminal : si deux branches vides la taille est 1 (car on a atteint une feuille)  
Le cas g√©n√©ral :   
-	si la branche de gauche est vide, on relance la fonction sur la branche de droite, ce qui signifie que l'on peut aller plus loin dans l'arbre donc on ajoute 1 √† la taille  
-	si la branche de droite est vide, on relance la fonction sur la branche de gauche, ce qui signifie que l'on peut aller plus loin dans l'arbre donc on ajoute 1 √† la taille  
-	si aucune des deux branches n'est vide, on relance la fonction sur les deux branches en r√©cup√©rant la taille des deux branches

### Parcours d'un arbre en largeur
Les parcours pr√©sent√©s pr√©c√©demment √©taient tous des parcours en profondeur. C‚Äôest-√†-dire que l'on parcourt l'arbre de haut en bas, puis vers la droite.  
Le parcours en largeur consiste √† parcourir tous les n≈ìuds situ√©s √† la m√™me profondeur de gauche √† droite avant de passer aux n≈ìuds de la profondeur suivante.  
Cet algorithme n'utilise pas de r√©cursivit√© mais utilise une structure de donn√©e : les files (FIFO)  

```
def parcoursLargeur_liste(arbre):
    file=[]
    file.append(arbre) 
    while file!=[]:
        arbre=file.pop(0)
        print(arbre[0],end='-') 
        if arbre[1]!=[]:
            file.append(arbre[1]) 
        if arbre[2]!=[]:
            file.append(arbre[2]) 
```
Au d√©but c'est l'arbre entier qui est dans la file.  

On d√©file l'arbre, on affiche sa racine puis on enfile le sous arbre droit s'il existe, on fait de m√™me avec le sous arbre gauche.  
On continue tant que la file n'est pas vide, c‚Äôest-√†-dire qu'il n'y a plus rien √† enfiler , on est arriv√© au bout de l'arbre.  





## Algorithmes sur les arbre binaire de recherche : ABR



Un arbre binaire de recherche est un cas particulier d'arbre binaire.  
Pour avoir un arbre binaire de recherche :    
‚Ä¢	il faut avoir un arbre binaire !  

‚Ä¢	il faut que les cl√©s de n≈ìuds composant l'arbre soient ordonnables (on doit pouvoir classer les n≈ìuds, par exemple, de la plus petite cl√© √† la plus grande)  

‚Ä¢	soit x un n≈ìud d'un arbre binaire de recherche. Si y est un n≈ìud du sous-arbre gauche de x, alors il faut que y.cl√© ‚â§ x.cl√©. Si y est un n≈ìud du sous-arbre droit de x, il faut alors que y.cl√© ‚â• x.cl√©    

### Recherche de cl√© dans un ABR


1er cas terminal : on arrive sur une branche vide, la valeur k n'a donc pas √©t√© trouv√© dans l'arbre  
2√®me cas terminal : la racine de l'arbre correspond √† la valeur k recherch√©e, on renvoie True  

Cas g√©n√©ral :  
‚Ä¢	si la valeur du n≈ìud k est < au n≈ìud examin√©e, on relance le test la branche gauche  
‚Ä¢	si la valeur du n≈ìud k est > au n≈ìud examin√©e, on relance le test la branche droite  

Cet algorithme de recherche d'une cl√© dans un arbre binaire de recherche ressemble beaucoup √† la recherche dichotomique.
En effet, on divise par deux la recherche en explorant soit le SAG soit le SAD.  
La complexit√© en temps dans le pire des cas de l'algorithme de recherche d'une cl√© dans un arbre binaire de recherche est $O(log_2(n))$, √† condition que l'arbre soit √©quilibr√©. C'est donc un algorithme rapide.
